```python
import pandas as pd
import matplotlib.pyplot as plt
df  = pd.read_csv("1_1d_lin.plt", delimiter=' ', names=['x', 'y'])
plt.plot(df['x'], df['y'], label='1D')
df  = pd.read_csv("1_2d_lin.plt", delimiter=' ', names=['x', 'y'])
plt.plot(df['x'], df['y'], label='2D')
df  = pd.read_csv("1_3d_lin.plt", delimiter=' ', names=['x', 'y'])
plt.plot(df['x'], df['y'], label='3D')
plt.legend()
plt.show()
```

```python
def foo(label):
    df  = pd.read_csv("3_3d_lin.plt_"+label, delimiter=' ', names=['x', 'y'])
    plt.plot(df['x'], df['y'], label=label)
for i in range(1, 5):
    foo(str(i))
plt.legend()
plt.show()
```
